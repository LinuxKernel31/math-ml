```python
from __future__ import absolute_import, division, print_function

__author__ = 'Robert John Navas'
__version__ = '1.0.0'

import matplotlib.pyplot as plt
import matplotlib.style as style
import numpy as np
import seaborn as sns
```

```python
style.use('seaborn-dark')
a = np.array([2,3])
b = np.array([4,-2])
plt.grid()
origin= [0],[0]

vectors = np.array([a,b])

plt.quiver(*origin, vectors[:,0], vectors[:,1], scale= 21, color=['b','g'])
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()


```

```python
vectors = np.array([a,b, a+b])
```

```python
origin= [0],[0]


plt.quiver(*origin, vectors[:,0], vectors[:,1], scale= 21, color=['r','b','g'])
plt.xlim(-7,7)
plt.ylim(-7,7)
plt.show()
```

```python
vectors = np.array([a,b, a*3, b*2])
```

```python
origin= [0],[0]


plt.quiver(*origin, vectors[:,0], vectors[:,1], scale= 21, color=['r','b','g','m'])
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.show()
```

```python
vectors = np.array([a,b,a-b,b-a])
plt.quiver(*origin, vectors[:,0], vectors[:,1], scale= 21, color=['r','b','g','m'])
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.show()
```
